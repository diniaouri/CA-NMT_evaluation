## Attention heatmap examples

```python
## Cherry picking 
def plot_heatmap(file_path):
    df = pd.read_csv(file_path, delimiter='\t')

    # Set the first column as the index
    df.set_index(df.columns[0], inplace=True)

    # Replace '.' with NaN
    df.replace('.', float('0.00'), inplace=True)

    # Convert DataFrame to numeric
    df = df.apply(pd.to_numeric, errors='coerce')

    plt.figure(figsize=(10, 8))
    sns.heatmap(df.transpose(), cmap='Blues', annot=True, fmt='.2f', cbar=True)  # Transpose the DataFrame
    plt.xticks(rotation=45)  # Remove rotation from x-axis labels
    plt.yticks(rotation=45)  # Rotate y-axis labels
    plt.show()

# Example usage:
file_paths = [
    '/home/user/Documents/GitHub/CA-NMT_evaluation/Multi-encoder_k3_model/normalise_trunc_point/normalise_trunc_point_26_1.tsv',
    '/home/user/Documents/GitHub/CA-NMT_evaluation/Multi-encoder_k3_model/normalise_trunc_point/normalise_trunc_point_54_1.tsv',
    '/home/user/Documents/GitHub/CA-NMT_evaluation/Multi-encoder_k3_model/normalise_trunc_point/normalise_trunc_point_40_1.tsv'
]

for path in file_paths:
    plot_heatmap(path)
```
